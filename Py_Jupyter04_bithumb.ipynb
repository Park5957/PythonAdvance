```python
#Building NFT
import sys #하드웨어 접근모듈
from PyQt5.QtWidgets import * #파이큐티 관련모듈
from PyQt5 import uic #UI 관련모듈
from PyQt5.QtCore import * #차이머 관련 모듈
import pybithumb
import pykorbit

form_class = uic.loadUiType("Py_Jupyter04_bithumb.ui")[0]
tickers = ["BTC","ETH","XRP","ADA"] # 조회를 원하는 BC 의 리스트를 생성

class Mywindow(QMainWindow,form_class): #MyWindow 클래스를 상속 받아서 작성
    def __init__(self):                 #생성자로 창 생성
        super().__init__()               # 부모의 생성자를 이용
        self.setupUi(self)               # ui 창을 생성
#         self.pushButton.clicked.connect(self.bnk_clicked)  #버튼 클릭시 연결되는 매서드 정의
        timer=QTimer(self) #타이머 객체생성
        timer.start(5000) #타이머로 시간이 지나면 조회가 진행 될 수 있도록 셋팅
        timer.timeout.connect(self.timeout)
      
        
   
    
    
    #5일 이동평균선으로 예를 들어보면 최근 5일간의 주가를 종가 기준으로 합산해서 5로 나눠 평균을 내는 것이다.
    def get_market_infos(self, ticker):
        df = pybithumb.get_ohlcv(ticker) #get_ohlcv() 해당 코인의 다양한 정보를 가져온다.
        ma5 = df["close"].rolling(window=5).mean() #종가들을 5일씩 평균을 전부 계산
        last_ma5 = ma5[-2] #가장 최근 5일치의 평균만을 추출
        price = pybithumb.get_current_price(ticker) #해당 코인의 현재가격
               
        
        state = None
 
        if price > last_ma5 :
            state = "상승장"
        else :
            state = "하락장"
            
        return price,last_ma5,state
        
    def timeout(self):
        for i,ticker in enumerate(tickers): #tickers 의 내용을 넘버링과 함께 각각 넘김
            item = QTableWidgetItem(ticker)
            self.tableWidget.setItem(i,0,item)

            
            price,last_ma5,state = self.get_market_infos(ticker)
            
            self.tableWidget.setItem(i,1,QTableWidgetItem(str(price)))
            self.tableWidget.setItem(i,2,QTableWidgetItem(str(last_ma5)))
            self.tableWidget.setItem(i,3,QTableWidgetItem(state))
    
    
class Mywindow2(QMainWindow,form_class):
    def __init__(self2):                 #생성자로 창 생성
        super().__init__()               # 부모의 생성자를 이용
        self2.setupUi(self2)
#         self.pushButton.clicked.connect(self.bnk_clicked)  #버튼 클릭시 연결되는 매서드 정의
        timer2=QTimer(self2) #타이머 객체생성
        timer2.start(5000) #타이머로 시간이 지나면 조회가 진행 될 수 있도록 셋팅
        timer2.timeout.connect(self2.timeout2)

    def get_market_infos2(self2, ticker):              
        df2 = pykorbit.get_ohlc(ticker) #get_ohlcv() 해당 코인의 다양한 정보를 가져온다.
        ma52 = df2["close"].rolling(window=5).mean() #종가들을 5일씩 평균을 전부 계산
        last_ma52 = ma52[-2] #가장 최근 5일치의 평균만을 추출
        price2 = pykorbit.get_current_price(ticker) #해당 코인의 현재가격
        
        state2 = None
 
        if price2 > last_ma52 :
            state2 = "상승장"
        else :
            state2 = "하락장"
            
        return price2,last_ma52,state2
 
    def timeout2(self2):
        for i,ticker in enumerate(tickers):
            item = QTableWidgetItem(ticker)
            self2.tableWidget_2.setItem(i,0,item)
          
            price2,last_ma52,state2 = self2.get_market_infos2(ticker)
            
            self2.tableWidget_2.setItem(i,1,QTableWidgetItem(str(price2)))
            self2.tableWidget_2.setItem(i,2,QTableWidgetItem(str(last_ma52)))
            self2.tableWidget_2.setItem(i,3,QTableWidgetItem(state2))    
            
            
            
    
app = QApplication(sys.argv)             # 파이썬은 원래 인터프리터 -> 한줄씩 실행후 종료
window = Mywindow()                      # 윈도우 클래스로 객체 생성
window2 = Mywindow2() 
window.show()                            # 생성한 객체를 통해 창을 보여주는
# window2.show() 
app.exec_()   
```
